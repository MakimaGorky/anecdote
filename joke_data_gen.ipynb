```python
import requests as rq

# html_text

start_page_number = 1
number_of_pages = 2021 + 1 # because there is 2021 pages in bashorg

html_texts = []

for num_page in range(start_page_number, number_of_pages):
    target_url = 'http://bashorg.org/page/%s/' % num_page
    print(target_url)
    html_texts.append(rq.get(target_url).text)

# html_texts

data_file = open('data.txt', 'w')

for html_text in html_texts:
    data_file.write(html_text)
```

```python
from bs4 import BeautifulSoup as bs

import codecs as cs

html_data = cs.open('data.txt', 'r', encoding = 'windows-1251', errors = 'ignore')

some_soup = bs(html_data, 'html.parser')

all_text = some_soup.get_text()

parsed_text = all_text.split('\n')

# parsed_text

anecdata = []

for cit_ind in range(len(parsed_text) - 2):
    if ('добавлено' in parsed_text[cit_ind]) and (parsed_text[cit_ind + 1] == ''):
        anecdata.append(parsed_text[cit_ind + 2])

# anecdata

file = cs.open('bash_of_jokes.txt', 'w', encoding='windows-1251', errors='ignore')

for anec in anecdata:
    file.write(anec + '\n')
```
